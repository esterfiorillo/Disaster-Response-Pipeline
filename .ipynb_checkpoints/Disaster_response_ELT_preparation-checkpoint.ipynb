```python
# import libraries
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
messages = pd.read_csv("/content/gdrive/My Drive/udacity/messages.csv")
messages.head()
```

```python
categories = pd.read_csv("/content/gdrive/My Drive/udacity/categories.csv")
messages.head()
```

```python
df = messages.merge(categories, how='inner', on='id')
df.head()
```

```python
categories = df['categories'].str.split(pat=';', expand=True)
names = list(categories.iloc[0])
names = [i.split('-')[0] for i in names]
categories.set_axis(names, axis=1,inplace=True)
```

```python
categories.head()
```

```python
for i in names:
  categories[i] = categories[i].str.slice(-1).astype(int)
```

```python
categories.head()
```

```python
df = df.drop(columns=['categories'])
df = pd.concat([df, categories], axis=1)
df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
engine = create_engine('sqlite:///ETL_Preparation.db')
df.to_sql('ETL_Preparation', engine, index=False)
```
